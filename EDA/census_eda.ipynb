```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
census_data = pd.read_csv('../../ci3_confounders/data_for_analysis/prepped_census/census_county_interpolated.csv')
```

```python
sub_df = census_data[['land_area', 'hispanic_pct', 'poverty', 'median_house_value', 'blk_pct',
       'white_pct', 'no_grad', 'median_household_income', 'owner_occupied', 'median_age','population_density']]
sns.pairplot(sub_df)
```

```python
# look at outliers, possible normalization, correlations
# positive trends - median household income and median household value; no grad and hispanic pct;
## poverty and hispanic pct; ower occupied and median age; poverty and no grad;
# negtive trends - poverty and median household income;
```

```python
plt.figure(figsize=(8,6))
plt.scatter(census_data['hispanic_pct'], census_data['no_grad'])
plt.xlabel("% hispanic")
plt.ylabel("% not graduating highschool")
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.scatter(census_data['median_household_income'], census_data['poverty'])
plt.ylabel("% poverty")
plt.xlabel("median household income")
plt.show()
```
